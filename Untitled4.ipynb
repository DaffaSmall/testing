```python
text = """This was a easy to read book that made me want to keep reading on and on, not easy to put down.It left me wanting to read the follow on, which I hope is coming soon. I used to read a lot but have gotten away from it. This book made me want to read again. Very enjoyable."""\
.strip().split(" ")
```

```python
X = dataset['train']['content'][:50000]
X_ = np.zeros([50000, 512])
for z, x in enumerate(X):
    print(z, "/", 50000)
    X_[z] = np.array([sum([x.pos_score() - x.neg_score()
    for x in list(swn.senti_synsets(y))])
    for y in text]).resize(512)
```
